```python
# Content type distribution
disney_hotstar['type'].value_counts().plot(kind='bar')
plt.title('Distribution of Content Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average IMDb ratings by content type
disney_hotstar.groupby('type')['imdb_rating'].mean()
```

```python
# Top 5 highest rated content
disney_hotstar.nlargest(5, 'imdb_rating')[['title', 'type', 'imdb_rating']]
```

```python
# Genre analysis (taking first genre for each title)
disney_hotstar['primary_genre'] = disney_hotstar['genre'].str.split(',').str[0]
plt.figure(figsize=(12, 6))
disney_hotstar['primary_genre'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Primary Genres')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Average runtime for movies
disney_hotstar[disney_hotstar['type'] == 'movie']['runtime'].str.extract('(\d+)').astype(float).mean()
```

```python
# Correlation analysis between numerical features
correlation_matrix = disney_hotstar[['metascore', 'imdb_rating']].corr()
correlation_matrix
```

```python
# Trend of content release over the years
disney_hotstar['year'] = disney_hotstar['year'].apply(lambda x: str(x).split('–')[0])  # Handle year ranges
release_trend = disney_hotstar['year'].value_counts().sort_index()
release_trend.plot(kind='line')
plt.title('Trend of Content Release Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.tight_layout()
plt.show()
```

```python
# Language distribution
plt.figure(figsize=(10, 5))
disney_hotstar['language'].str.split(',').str[0].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Primary Languages')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Rating distribution
plt.figure(figsize=(10, 5))
disney_hotstar['rated'].value_counts().plot(kind='bar')
plt.title('Content Rating Distribution')
plt.tight_layout()
plt.show()
```

```python
# Summary statistics
print('\Content Summary:')
print('Total number of titles:', len(disney_hotstar))
print('Number of unique directors:', disney_hotstar['director'].nunique())
print('Average IMDb rating:', round(disney_hotstar['imdb_rating'].mean(), 2))
print('Percentage of content with awards:', round((disney_hotstar['awards'].notna().sum() / len(disney_hotstar)) * 100, 2), '%')
```

```python
# Content by country
plt.figure(figsize=(10, 5))
disney_hotstar['country'].str.split(',').str[0].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Production Countries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
